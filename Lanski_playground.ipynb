## Workflow for Jupiter Notebooks:
- Modal Navigation!
    - Press Enter to enter edit, press Esc to enter navigation mode
- While in Navigation mode:
    - J to go down, K to go up
    - Press M on a cell to transfrom it to a Markdown
    - Press Y to transform it back to code
    - Press Shift+Enter to run a cell
    - Creating and deleting cells by pressing `a` to create a cell above and `b` to create it below. 
    - Press `dd` to delete a cell
    - Shift+navigating cells highlights them. 
        - Pressing shift+m over highlighted cells merges them
